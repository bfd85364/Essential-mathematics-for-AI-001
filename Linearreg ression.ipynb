```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
%matplotlib inline
import seaborn as sn

df = pd.read_csv('House_Rent_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['Posted On', 'Area Locality', 'Floor'], axis=1)
df.head()

df = pd.get_dummies(df, columns=['Area Type', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact'])
df.head()
```

```python
X = df
Y = df['Rent']
X.drop('Rent', axis = 1, inplace = True)

std = StandardScaler()
X = std.fit_transform(X)

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.3, random_state = 5)

model = LinearRegression()
model.fit(xtrain, ytrain)

pred = model.predict(xtest)

MSE = mean_squared_error(ytest, pred)
print("The mean squared error MSE is:", MSE)

MAE = mean_absolute_error(ytest, pred)
print("The mean absolute error MAE is:", MAE)
```
